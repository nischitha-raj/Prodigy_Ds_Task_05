```python
# Print column names of df_subset
print(df_subset.columns)

# Extract relevant columns for analysis
relevant_columns = ['Time', 'Day_of_week', 'Weather_conditions', 'Road_surface_conditions']
df_subset = df[relevant_columns]


# Count accidents by road conditions
road_condition_counts = df_subset['Road_surface_conditions'].value_counts()
```

```python
import matplotlib.pyplot as plt
# Visualize accident hotspots and contributing factors
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
road_condition_counts.plot(kind='bar', color='skyblue')
plt.title('Accidents by Road Conditions')
plt.xlabel('Road Conditions')
plt.ylabel('Number of Accidents')
```

```python
weather_counts = df_subset['Weather_conditions'].value_counts()

# Count accidents by time of the day
time_of_day_counts = df_subset['Time'].apply(lambda x: x.split(':')[0]).value_counts()

# Visualize accident hotspots and contributing factors
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
road_condition_counts.plot(kind='bar', color='skyblue')
plt.title('Accidents by Road Conditions')
plt.xlabel('Road Conditions')
plt.ylabel('Number of Accidents')
```

```python
time_of_day_counts = df_subset['Time'].apply(lambda x: x.split(':')[0]).value_counts()

# Visualize accident hotspots and contributing factors
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
road_condition_counts.plot(kind='bar', color='skyblue')
plt.title('Accidents by Road Conditions')
plt.xlabel('Road Conditions')
plt.ylabel('Number of Accidents')

plt.subplot(1, 3, 2)
weather_counts.plot(kind='bar', color='lightgreen')
plt.title('Accidents by Weather')
plt.xlabel('Weather')
plt.ylabel('Number of Accidents')

plt.subplot(1, 3, 3)
time_of_day_counts.sort_index().plot(kind='bar', color='salmon')
plt.title('Accidents by Time of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')

plt.tight_layout()
plt.show()
```
